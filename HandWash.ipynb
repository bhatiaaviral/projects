Frame Capture function captures BGR images for both training and validation sets

```
def FrameCapture(pathd,paths,img_count): 
      
    # Path to video file 
    
    vidObj = cv2.VideoCapture(paths) 
    
    # Used as counter variable 
    
  
    # checks whether frames were extracted 
    success = 1
    count=1
    while success: 
        
        # vidObj object calls read 
        # function extract frames 
        success, image = vidObj.read() 
        
        # Saves the frames with frame-count 
        if count%10==0:
            print('here')
            #print(pathd)
            path=os.path.join(pathd,'image{}.jpg'.format(img_count))
            #print('here  '+path)  
            if success: 
                cv2.imwrite(path, image) 
            img_count += 1
        count += 1
    return img_count
  
```

RGBtoGRAY function converts RGB images to grayscale images

```
def RGBtoGRAY(pathd,paths,img_count):

    image = cv2.imread(paths)
    RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
    gray = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
    path=os.path.join(pathd,'image{}.jpg'.format(img_count))
    cv2.imwrite(path, gray)
    img_count += 1
    return img_count
```

```
from keras.layers import Input, Lambda, Dense, Flatten, Dropout
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```
IMAGE_SIZE = [224, 224]

train_path = '/content/drive/My Drive/HW_image_dataset'
valid_path = '/content/drive/My Drive/val'
```

```
vgg = VGG16(input_shape=IMAGE_SIZE+[3], weights='imagenet', include_top=False)

# don't train existing weights
for layer in vgg.layers:
  layer.trainable = False
```

```
x=Flatten()(vgg.output)
x=Dropout(0.5)(x)
x=Dense(4096,activation='relu')(x)
x=Dropout(0.5)(x)
pred=Dense(12,activation='softmax')(x)
model = Model(inputs=vgg.input, outputs=pred)
```

```
model.summary()
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory(train_path, target_size = (224, 224), batch_size = 32, class_mode = 'categorical')

test_set = test_datagen.flow_from_directory(valid_path, target_size = (224, 224), batch_size = 32, class_mode = 'categorical')
```

```
import tensorflow as tf
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.models import load_model

model = load_model("/content/drive/My Drive/vgg16_3.h5")

checkpoint = ModelCheckpoint("/content/drive/My Drive/vgg16_3.h5", monitor='val_accuracy', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)

early = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=20, verbose=1, mode='auto')

r = model.fit_generator( training_set, validation_data=test_set, epochs=5, steps_per_epoch=len(training_set), validation_steps=len(test_set), callbacks=[checkpoint,early])
#model.save('/content/drive/My Drive/vggtransfer.h5')
```
